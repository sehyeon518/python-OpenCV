# 예제 5.6 히스토그램 계산 4: 컬러 영상의 2채널 히스토그램

```python
import cv2
import matplotlib.pyplot as plt
```

```python
bgr = cv2.imread('../sample/lena.png')

hist01 = cv2.calcHist(images=[bgr], channels=[0,1], mask=None, histSize=[32,32], ranges=[0,256,0,256])
hist01
```

```python
plt.title('hist01')
plt.ylim(0,31)
plt.imshow(hist01, interpolation="nearest")
plt.show()
```

```python
hist02 = cv2.calcHist(images=[bgr], channels=[0,2], mask=None, histSize=[32,32], ranges=[0,256,0,256])
plt.title('hist02')
plt.ylim(0,31)
plt.imshow(hist02, interpolation="nearest")
plt.show()
```

```python
hist12 = cv2.calcHist([bgr], [1,2], None, [32,32], [0,256,0,256])
plt.title('hist12')
plt.ylim(0,31)
plt.imshow(hist02, interpolation="nearest")
plt.show()
```
