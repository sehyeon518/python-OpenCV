# 예제 5.5 히스토그램 계산 3: 컬러 영상

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

src = cv2.imread('../sample/lena.png')
```

```python
histColor = ('b', 'g', 'r')
for i in range(3):
    hist = cv2.calcHist(images=[src], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(hist, color=histColor[i])
plt.show()
```
