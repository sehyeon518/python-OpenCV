```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('sample/ball.mp4')

# height, width = (int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)), int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)))
```

```python
# 검출된 공의 중심 좌표와 반지름
CIRCLE_X = 0
CIRCLE_Y = 0
RADIUS = 0
```

```python
while True:
    retval, src = cap.read()
    if not retval:
        break
    
    blur = cv2.GaussianBlur(src, (9,9), 0)
    gray = cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)

    circles = cv2.HoughCircles(gray, method=cv2.HOUGH_GRADIENT, dp=1, minDist=200, param1=250, param2=15, minRadius=30, maxRadius=200)
    if circles is not None:
        circles = np.int32(circles)

        # 검출된 원이 한개인 경우 공을 검출한 것으로 간주
        if len(circles[0,:]) == 1:
            circle = circles[0,:][0]
            CIRCLE_X, CIRCLE_Y, RADIUS = circle
        # 다수의 원이 검출되었을 경우 이전에 감지된 원과 중심 좌표가 가장 근접한 원을 공으로 간주함
        else:
            distinction = 10000
            for circle in circles[0,:]:
                cx, cy, r = circle
                tmp = (CIRCLE_X - cx)**2 + (CIRCLE_Y - cy)**2
                if tmp < distinction:
                    CIRCLE_X, CIRCLE_Y, RADIUS = circle
        cv2.circle(src, (CIRCLE_X, CIRCLE_Y), RADIUS, (0,255,0), 2)
        print(CIRCLE_X, CIRCLE_Y, RADIUS)


    cv2.imshow('src', src)
        
    key = cv2.waitKey(5)
    if key == 27:
        break

cv2.destroyAllWindows()
```
