# 예제 2.5 Matplotlib 3: 여백 조정 및 영상 저장

```python
import cv2
from matplotlib import pyplot as plt
```

```python
imageFile = '../sample/why.jpeg'
imgGray = cv2.imread(imageFile, cv2.IMREAD_GRAYSCALE)
```

### <span style='background-color:#fff5b1'>plt.figure(figsize = (x, y))</span>
크기를 (x인치, y인치)로 설정한다

```python
plt.figure(figsize=(6,6))
```

### <span style='background-color:#fff5b1'>plt.subplots_adjust(left=0, right=1, bottom=0, top=1)</span>
영상 출력 범위를 좌우[0,1], 상하[0,1]로 조정한다\
이때 left < right, bottom < top이어야 한다

### <span style='background-color:#fff5b1'>plt.savefig(dir)</span>
파일 저장

```python
import os

plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
plt.imshow(imgGray, cmap='gray')
plt.axis('off')
os.mkdir('../sample/0205')
plt.savefig('../sample/0205/0205.png')
plt.show()
```
