# 예제 2.1 영상 파일 읽기 및 화면표시

```python
import cv2
import numpy as np
```

### <span style='background-color:#fff5b1'>cv2.imread()</span> 영상 파일을 numpy.ndarray의 배열로 읽어 반환
읽기에 실패하면 none을 return

```python
imageFile = '../sample/why.jpeg' # 영상 파일을 numpy.ndarray의 배열로 읽어 반환
img = cv2.imread(imageFile)
img2 = cv2.imread(imageFile, 0)
print('img shape=', img.shape)
img[0]
```

### <span style='background-color:#fff5b1'>cv2.write(filename, img)</span> numy.ndarray의 배열 img를 filename의 영상파일로 저장

### <span style='background-color:#fff5b1'>cv2.imshow(window, img)</span> 윈도우 생성하여 영상 표시
window에 img 표시
### <span style='background-color:#fff5b1'>cv2.waitKey(delay)</span> 키보드 입력 대기
키보드 입력을 delay milli seconds 동안 대기\
default=0 -> 무한대기 -> 윈도우가 없어지지 않게 한다\
누른 키의 번호를 반환한다
### <span style='background-color:#fff5b1'>cv2.destroyAllWindows()</span> 모든 윈도우 파괴

```python
cv2.imshow('sample color', img)
cv2.imshow('sample grayscale', img2)

cv2.waitKey()

cv2.destroyAllWindows()
```
