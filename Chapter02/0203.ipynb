# 예제 2.3 Matplotlib 1: 컬러 영상 표시

```python
import cv2
from matplotlib import pyplot as plt
```

```python
imageFile = '../sample/why.jpeg'
imgBGR = cv2.imread(imageFile)
```

### <span style='background-color:#fff5b1'>plt.axis()</span> 축 표시 범위 설정
**plt.axis((ximin, xmax, ymin, ymax))** \
**plt.axis(option)**\
'on' 축 표시\
'off' 축 표시 안함\
'equal' 두 축의 범위와 스케일을 동일하게 표시\
'scaled' 두 축을 같은 크기로 표시, 값이 없는 부분은 잘라서 표시\
'tight' 모든 데이터를 볼 수 있을 정도로 축의 범위를 크게 설정\
'auto' 축의 스케일을 자동으로 설정\
'normal' == 'auto'\
'image' 데이터 범위에 대해 축의 범위를 사용한 'scaled'\
'square' 두 축을 같은 크기로 표시

### <span style='background-color:#fff5b1'>plt.imshow()</span>

### <span style='background-color:#fff5b1'>plt.show()</span>
영상의 채널 순서 때문에 영상의 컬러가 다르게 표시된다\
OpenCV로 읽은 컬러 영상 채널 순서는 BGR, Matplotlib는 RGB 순서

```python
plt.axis('off')
plt.imshow(imgBGR)
plt.show()
```

```python
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(imgRGB)
plt.show()
```
