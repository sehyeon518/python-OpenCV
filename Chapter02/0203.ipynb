# 예제 2.3 Matplotlib 1: 컬러 영상 표시

```python
import cv2
from matplotlib import pyplot as plt
```

### plt.axis('off) X, Y축을 표시하지 않는다
### plt.imshow() imgBGR 출력
### plt.show() 영상의 채널 순서 때문에 영상의 컬러가 다르게 표시된다
OpenCV로 읽은 컬러 영상 채널 순서는 BGR
Maatplotlib는 RGB 순서

```python
imageFile = '/home/sehyeon/사진/이세현.jpg'
imgBGR = cv2.imread(imageFile)
plt.axis('off')
plt.imshow(imgBGR)
plt.show()
```

```python
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
plt.show()
```
