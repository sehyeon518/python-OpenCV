# 예제 2.6 Matplotlib 4: 서브플롯에 영상 표시

```python
import cv2
from matplotlib import pyplot as plt
```

```python
imgRGB1 = cv2.imread('/home/sehyeon/사진/이세현.jpg')
imgRGB2 = cv2.imread('/home/sehyeon/사진/배경 이미지/Screenshot_20221008_110206.jpg')
```

### 컬러 변환: BGR -> RGB

```python
imgRGB1 = cv2.cvtColor(imgRGB1, cv2.COLOR_BGR2RGB)
imgRGB2 = cv2.cvtColor(imgRGB2, cv2.COLOR_BGR2RGB)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(5, 5), sharey = True)
fig.canvas.manager.set_window_title('Sample Pictures')

ax[0].axis('off')
ax[1].axis('off')

ax[0].imshow(imgRGB1, aspect='auto')
ax[1].imshow(imgRGB2, aspect='auto')

plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.05, hspace=0.05)
plt.show()
```
