# 예제 4.21 cv2.randu()에 2차원 균등 분포 난수 좌표

```python
import cv2
import numpy as np
import time
```

```python
dst = np.full((512, 512, 3), (255, 255, 255), dtype = np.uint8) # (255, 255, 3)을 512*512에 채운다
nPoints = 100
dst
```

```python
pts = np.zeros((1, nPoints, 2), dtype = np.uint16)
pts
```

```python
pts.shape
```

### <span style='background-color:#fff5b1'>cv2.setRNGSeed(int(time.time())</span>
난수 생성 초기화

```python
cv2.setRNGSeed(int(time.time()))
```

```python
cv2.randu(pts, (0,0), (512,512))
```

```python
for k in range(nPoints):
    x, y = pts[0, k][:]
    cv2.circle(dst, (x,y), radius = 5, color = (0,0,255), thickness = -1)

cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
